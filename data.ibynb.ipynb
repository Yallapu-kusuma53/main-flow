Filtering Rows:

->Filter by a condition:
Example: Get users with app usage time greater than 200 minutes per day.


```python
# Convert 'App Usage Time (min/day)' to numeric
df['App Usage Time (min/day)'] = pd.to_numeric(df['App Usage Time (min/day)'], errors='coerce')

# Now, apply the condition to filter the rows
filtered_df = df[df['App Usage Time (min/day)'] > 200]
filtered_df.head()
```

Filter by multiple conditions:

Example: Get users who use Android devices and have battery drain higher than 1500 mAh/day.

```python
# Convert the 'Battery Drain (mAh/day)' column to numeric
df['Battery Drain (mAh/day)'] = pd.to_numeric(df['Battery Drain (mAh/day)'], errors='coerce')

# Apply the filtering condition for Android devices with battery drain > 1500 mAh/day
filtered_df_battery = df[(df['Operating System'] == 'Android') & (df['Battery Drain (mAh/day)'] > 1500)]

# Display the filtered data
filtered_df_battery.head()
```

Group By and Aggregation:

->Group by a column and aggregate:
Example: Group by Operating System and get the mean app usage time.

```python
df.groupby('Operating System')['App Usage Time (min/day)'].mean()
```

Sorting Data:

->sort values by a column:
Example: Sort users by app usage time in descending order.

```python
df.sort_values(by='App Usage Time (min/day)', ascending=False)
```

Adding New Columns:

->Create a new column:
Example: Create a column for app usage in hours.

```python
df['App Usage Time (hours/day)'] = df['App Usage Time (min/day)'] / 60
```

Removing or Renaming Columns;

->Remove a column:
Example: Drop the Age column.

```python
df.drop('Age', axis=1, inplace=False)  # inplace=True for permanent drop
```

Rename a column:
Example: Rename App Usage Time (min/day) to App Usage (min/day).

```python
df.rename(columns={'App Usage Time (min/day)': 'App Usage (min/day)'}, inplace=True)
df.head()
```

Handling Missing Data
Find missing values:


```python
df.isnull().sum()
```

Pivot Table:

->Create a pivot table:
Example: Create a pivot table showing average battery drain by Operating System and Gender.

```python
df.pivot_table(values='Battery Drain (mAh/day)', index='Operating System', columns='Gender', aggfunc='mean')
```
